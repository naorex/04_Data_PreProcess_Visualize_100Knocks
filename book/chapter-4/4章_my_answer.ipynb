## ノック５１：テキストファイルを読み込もう

```python
with open("data/hashire_merosu.txt", mode="r", encoding="shift-jis") as f:
    content = f.read()
print(content)
```

## ノック５２：本文を抽出して１つに纏めよう

```python
# 改行を削除
content = " ".join(content.split())
content
```

```python
# 文字の正規化
import unicodedata
content = unicodedata.normalize("NFKC", content)
content
```

```python
import re
pattern = re.compile(r'^.+(#地から1字上げ].+#地から1字上げ]).+$')
body = re.match(pattern, content).group(1)
print(body)
```

```python
body = body.replace('#地から1字上げ] ------------------------------------------------------- ', '')
body = body.replace(' [#地から1字上げ]', '')
body
```

## ノック５３：本文以外の項目を取り出そう

```python
# 先頭2行を抽出
with open('data/hashire_merosu.txt', mode='r', encoding='shift-jis') as f:
  title = f.readline()
  author = f.readline()
print(title)
print(author)
```

```python
title = title.replace('\n', '')
print(title)
author = author.replace('\n', '')
print(author)
```

```python
with open('data/hashire_merosu.txt', mode='r', encoding='shift-jis') as f:
  content = f.readlines()
content
```

```python
# dataframe に格納
import pandas as pd
df = pd.DataFrame(content, columns=['text'])
df['text'] = df['text'].str.replace('\n', '')
df
```

```python
date = df[(df['text'].str.contains('日公開')
           )|(df['text'].str.contains('日修正'))].copy()
print(date)

date['text'] = date['text'].str.replace('公開', '')
date['text'] = date['text'].str.replace('修正', '')
print(date)

date['text'] = date['text'].str.replace('年', '/')
date['text'] = date['text'].str.replace('月', '/')
date['text'] = date['text'].str.replace('日', '')
print(date)

date['text'] = pd.to_datetime(date['text'])
print(date)
date.dtypes
```

```python
release_date = date.iat[0, 0] # 特定の1つの要素へのアクセスは .iloc より .iat が高速
update_date = date.iat[1, 0]
print(release_date)
print(update_date)
date = update_date - release_date
print(date)
```

```python
booklist = pd.DataFrame([[title, author, release_date, update_date, body]],
                        columns=['title', 'author', 'release_date', 'update_date', 'body'])
booklist
```

## ノック５４：形態素解析で単語に分割しよう

以下のライブラリを用いて実行
- pip install mecab-python3 unidic-lite

```python
# 単語の先頭4文字
import MeCab
tagger = MeCab.Tagger()
body = booklist.iloc[0, 4]
parsed = tagger.parse(body).split('\n')
parsed[:4]
```

```python
parsed[-4:]
```

```python
# 末尾2つは不要なので削除
parsed = parsed[:-2]
parsed[-4:]
```

## ノック５５：分割した単語をデータフレームで保持しよう


```python
# Mecab 実行結果を多次元リストに格納
# 解説: parsed の各文字列 s に対して、「タブまたはカンマで分割する」という処理を行い、その結果（分割された文字列のリスト）を順番に生成するイテレータを返す処理。*values と指定する事で、値をひとつづつ持てる。
*values, = map(lambda s: re.split(r'\t|,', s), parsed)
values[:4]
```

```python
# dataframe へ格納していく
import pandas as pd
columns = ['表層形', '発音', '原形', '読み', '品詞', '品詞細分類1', '品詞細分類2', '品詞細分類3', '活用型', '活用形']
mecab_df = pd.DataFrame(data=values, columns=columns)
mecab_df['大品詞'] = [y[0] for y in mecab_df['品詞'].str.split('-')]

print(len(mecab_df))
mecab_df.head(4)
```

## ノック５６：名詞と動詞を取り出そう

```python
print(mecab_df.groupby(['大品詞']).size().sort_values(ascending=False))
```

```python
# 名詞のみ
noun = mecab_df.loc[(mecab_df['大品詞'] == '名詞')]
noun
```

```python
# 名詞と動詞
verb = mecab_df.loc[(mecab_df['大品詞'] == '名詞')|(mecab_df['大品詞'] == '動詞')]
verb
```

## ノック５７：不要な単語を除外しよう

```python
# stop word を読み込み
with open('data/stop_words.txt', mode='r', encoding='utf-8') as f:
  stop_words = f.read().split()
stop_words
```

```python
print(len(noun))
noun = noun.loc[~noun['表層形'].isin(stop_words)]
print(len(noun))
display(noun.head())

print(len(verb))
verb = verb.loc[~verb['表層形'].isin(stop_words)]
print(len(verb))
display(verb.head())
```

## ノック５８：単語の使用状況をグラフで可視化しよう

```python
count = noun.groupby('表層形').size().sort_values(ascending=False)
count.name = 'count'
count = count.reset_index().head(10)
count
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib_fontja
plt.figure(figsize=(10, 5))
sns.barplot(x=count['count'], y=count['表層形'])
```

## ノック６０：n-gramを作ってみよう

```python
# 値の出現回数をカウント
import collections
counter = collections.Counter(bigram)
print(counter)
```
